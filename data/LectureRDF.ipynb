# Première lecture fichier RDF

## Import fichier RDF

```python
from rdflib.graph import Graph
#https://rdflib.readthedocs.io/en/3.4.0/intro_to_graphs.html#graph-methods-for-accessing-triples
```

```python
file_path='../data/LIFRANUM.rdf'
g = Graph()
g.parse(file_path)
```

```python
len(g) # print length > on a 1658 entités
```

 ## PARCOUR DU FICHIER
 
> les rdf représentent des tuples de trois entités : un sujet (ce qui est décrit), un prédicat (sous quel aspet celui-ci est décrit) et un objet (comment est il décrit)

```python
pred_list=[]
for subj, pred, obj in g:
    print("------------------")
    print("S:"+subj)
    print("P:"+pred)
    print("O:"+obj)
    if(pred not in pred_list):
        pred_list.append(pred)
    if (subj, pred, obj) not in g:
       raise Exception("It better be!")
```

## PARSING DES URL ET RECUPERATION DES DOMAINES
# création de dictionnaires : 
* pour le comptage
* pour avoir la liste des urls par domaines
* pour avoir les informations renseignée par domaines

```python
from urllib import parse

count_by_domain={}
url_by_domain={}
meta_by_domain={}
for subj, pred, obj in g:
    url=parse.urlsplit(subj)
    p=pred.replace("http://purl.org/dc/elements/1.1/","")
    if url.netloc not in meta_by_domain.keys():
        meta_by_domain[url.netloc]={}
    if p not in meta_by_domain[url.netloc].keys():
        meta_by_domain[url.netloc][p]=[]
    if obj not in meta_by_domain[url.netloc][p]:
        meta_by_domain[url.netloc][p].append(str(obj))  
    if(url.netloc in count_by_domain.keys()):
        if str(subj) is not None:
            url_by_domain[url.netloc].append(str(subj))
        count_by_domain[url.netloc]=count_by_domain[url.netloc]+1
        
    else:
        
        url_by_domain[url.netloc]={}
        if str(subj) is not None:
            url_by_domain[url.netloc]=[str(subj)]
        count_by_domain[url.netloc]=1
```

### Tri du dictionnaire de comptage et affichage du Top 20

```python
import operator
sorted_x = sorted(count_by_domain.items(), key=operator.itemgetter(1),reverse=True)
print(sorted_x[0:20])
```
