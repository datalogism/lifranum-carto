```python

import json
from bs4 import BeautifulSoup as bs

url="http://ile-en-ile.org/lit-haitienne/?fbclid=IwAR3EXGOyrU3djOqD04Nn61UKwtWyI0x7x79cTRWqL9ODT-LWYxEfW00EinY"

import requests
r  = requests.get(url)
data_c = r.text
data_c
```

```python
soup = bs(data_c)
data_authors={}
home=soup.find("div", {'class':'home-main'})
links=home.find("div", {'class':'nested-menu'})
for l in links:
    found=l.find("a")
    if(found != -1):
        data_authors[found.get_text()]={"url":found["href"]}
    
```

```python
     
    
file='C:/Users/Celian/Desktop/M2 HUMANUM/PROJET/lifranum_carto/data/ile_en_ile0.json'

with open(file, 'w', encoding='utf-8') as f:
    json.dump(data_authors, f, ensure_ascii=False, indent=4)
        
```

```python

for k in data_authors.keys():
    print(k)
    r  = requests.get(data_authors[k]["url"])
    data_c = r.text
    soup = bs(data_c)
    content=soup.find("section",{"class":"post_content"})
    bio=[]
    bio_elements=content.findAll("p",{"class":None})
    for el in bio_elements:
        bio.append(el.get_text())
    data_authors[k]["bio"]=bio
    production={}
    prods=content.findAll("h3")
    for p in prods:
        production[p.get_text().replace(":","")]=p.next_sibling.next_sibling
    data_authors[k]["productions"]=production
```

```python
     
    
file='C:/Users/Celian/Desktop/M2 HUMANUM/PROJET/lifranum_carto/data/ile_en_ile0.json'

with open(file, 'w', encoding='utf-8') as f:
    json.dump(data_authors, f, ensure_ascii=False, indent=4)
```
